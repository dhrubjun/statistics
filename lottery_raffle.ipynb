```python
import random
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
total = 0
total1 = []
total_avg = []
```

```python
for i in range(100):
  first_prize = random.sample([i for i in range(1,2500)],1)
  second_prize = random.sample([i for i in range(1,2500) if i not in first_prize],10)
  third_prize = random.sample([i for i in range(1,2500) if i not in first_prize and second_prize], 30)

  number = random.choice(range(1,2500))
  if number in third_prize:
    total = total + 100
    total_avg.append(total/(i+1))
    total1.append(total)

  elif number in second_prize:
    total = total + 500
    total_avg.append(total/(i+1))
    total1.append(total)

  elif number in first_prize:
    total = total + 8000
    total_avg.append(total/(i+1))
    total1.append(total)

  else :
    total = total - 8
    total_avg.append(total/(i+1))
    total1.append(total)

```

```python
figure(figsize=(10,7))
plt.plot(total1)
plt.xlabel('No. of game played')
plt.ylabel('Total winning')
```

```python
figure(figsize=(10,7))
plt.plot(total_avg)
plt.xlabel('No. of game played')
plt.ylabel('Average winning per game')
```
